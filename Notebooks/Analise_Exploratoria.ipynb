```python
# Importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
# Upload do arquivo
from google.colab import files
arquivo = files.upload()
```

```python
# Cria√ß√£o do Dataframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
# Visualizando as 5 primeiras linhas
df.head()
```

```python
# Quantidade de linhas e colunas
df.shape
```

```python
# Verificando o tipo dos dados
df.dtypes
```

```python
# Qual a 'Receita total'?
print(f' O Valor da Receita √© \033[4;32mR$ {df["Valor Venda"].sum().round(2)}\033[m')
```

```python
# Criando a coluna 'Custo'
df["Custo"] = df["Custo Unit√°rio"].mul(df["Quantidade"])
df.head()
```

```python
# Qual o 'Custo total'?
print(f'O custo total √© \033[4;32mR$ {df["Custo"].sum():,.2f}\033[m')  
```

```python
# Agora temos 'Receita', 'Custo Total'. Podemos achar 'Lucro Total'!
# Criar uma coluna 'Lucro' que ser√° ('Receita' - 'Custo')
df["Lucro"] = df["Valor Venda"] - df["Custo"]
df.head()
```

```python
# Total Lucro?
print(f'O lucro total √© \033[4;32mR$ {df["Lucro"].sum():,.2f}\033[m')
```

```python
# Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
df.head()
```

#### <font color=00C0C0>Precisamos converter a coluna 'Tempo_envio' para num√©rica. Queremos saber a m√©dia do tempo de envio para cada 'Marca'</font>

```python
# Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head()
```

```python
# Verificando o tipo da coluna 'Tempo_envio'
df["Tempo_envio"].dtype
```

```python
# M√©dia do tempo de envio por 'Marca'
media_tempo_envio_marca = df.groupby("Marca")["Tempo_envio"].mean().round(2).reset_index()
media_tempo_envio_marca
```

#### <font color=00C0C0>Missing Values</font>

```python
# Verificando exist√™ncia de dados faltantes
df.isnull().sum()
```

<font color=00C0C0>Lucro por 'Ano' e 'Marca'</font>

```python
pd.options.display.float_format = '{:20,.2f}'.format
```

```python
# Vamos Agrupar por 'Ano' e 'Marca'
# Resetando 'index'
lucro_por_ano_marca = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_por_ano_marca
```

```python
# Qual uantidade dos produtos vendidos por produto?
# Ao resetar o indice o gr√°fico n√£o aparece nome do produto e sim os n√∫meros.
total_produtos_vendidos = pd.DataFrame(df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = False)).reset_index()
total_produtos_vendidos
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending = True).plot.barh(title = "Total Produtos Vendidos", figsize=(10, 6), color = "#80C040");
```

```python
# Total lucro por ano
total_lucro_ano = df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().reset_index()
total_lucro_ano 
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title = "Lucro x Ano", figsize=(10, 6), color = "#C040C0")
plt.xlabel("Ano")
plt.ylabel("Receita")
plt.xticks(rotation = "horizontal");
```

```python
# Selecionando apenas as vendas de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title = "Lucro x M√™s", figsize=(10, 6), color = "#C040C0")
plt.xlabel("M√™s")
plt.ylabel("Lucro");
```

```python
df_2009.groupby(df_2009["Marca"])["Lucro"].sum().plot.bar(title = "Lucro x Marca", figsize=(10, 6), color = "#0000FF")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```

```python
df_2009.groupby(df_2009["Classe"])["Lucro"].sum().plot.bar(title = "Lucro x Classe", figsize=(10, 6), color = "#FF4080")
plt.xlabel("M√™s")
plt.ylabel("Lucro")
plt.xticks(rotation = "horizontal");
```
